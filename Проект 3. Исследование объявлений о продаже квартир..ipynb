### Этаж квартиры.

Нам нужно создать дополнительны столбец, в котором будет содержаться информация об этажности квартиры. Информация нужна в формате: первый, последний, другой этаж. Значит нам необходимо прибегнуть к категоризации данных, которые содержаться в столбце floor (этаж) и floors_total (количество этажей в доме). Если floor равен 1, то мы отнесем его к категории "первый", если floor равен floors_total, то этот этаж "последний", все прочие - "другой этаж". Очевидно, нам нужна функция. Назовем её category, а новый столбец назовём floor_category:

```python
def category (row):
    floor=row['floor']
    floors_total=row['floors_total']
    if floor==1:
        return 'первый'
    if floor==floors_total:
        return 'последний'
    else:
        return 'другой этаж'
propertys['floor_category'] = propertys.apply(category, axis=1)
propertys.head(10)
```

Столбец с категоризацией этажности мы добавили.
